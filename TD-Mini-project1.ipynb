```python
import pandas as pd
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.graph_objs as go
init_notebook_mode(connected=True)


file = pd.read_csv('userssharedsdfschoolimprovement2010grants.csv')

file
```

```python
file.describe()
```

```python
file.rename(columns= {'2010/11/Award Amount': 'Amount'}, inplace=True)
```

```python
file.columns
```

```python
file['Amount'] = file['Amount'].str.replace('$','')
file.head()
```

```python
file['Amount'] = file.Amount.astype(float)
```

```python
file.dtypes
```

```python
file['Street'], file['CityState'], file['Coordinates'] = file['Location'].str.split('\n').str

file['Latitude'], file['Longitude'] =  file['Coordinates'].str.replace('(','').str.replace(')','').str.split(',').str

file['ZIP'] = file['CityState'].str[-5:]

del file['Location']
del file['Coordinates']
del file['CityState']
```

```python
print (file.head(10))
```

```python
print(file.dtypes)
```

```python
file['ZIP'] = file['ZIP'].astype('int')
file['Latitude'] = file['Latitude'].astype('float64')
file['Longitude'] = file['Longitude'].astype('float64')

print(file.dtypes)
file.describe()
```

```python
colors = ["rgb(0,116,217)","rgb(255,65,54)","rgb(133,20,75)"
          ,"rgb(255,133,27)","lightgrey"] 
cities = []
scale = 10000
file['text'] = file['School Name'] + ' - ' + file['Amount'].astype(str)

limits = [(0,50),(60,100),(110,150),(160,200),(210,2000000)]

for i in range(len(limits)):
    lim = limits[i]
    df_sub = file[lim[0]:lim[1]]
    city =  dict(
        type = 'scattergeo',
        locationmode = 'USA-states',
        lon = df_sub['Longitude'],
        lat = df_sub['Latitude'],
        text = df_sub['text'],
        marker = dict(
            size = df_sub['Amount']/scale,
           # sizeref = 2. * max(df_sub['Amount']/scale) / (25 ** 2),
            colr = colors[i],
            line = dict(width=0.5, color='rgb(40,40,40)'),
            sizemode = 'area'
            ),
           name = '{0} - {1}'.format(lim[0],lim[1])  )
    cities.append(city)      
        

layout = dict(
        title = 'Schools That Received Awards',
        showlegend = True,
        geo = dict(
            scope='usa',
            projection=dict( type='albers usa' ),
            showland = True,
            landcolor = "rgb(217, 217, 217)",
            subunitcolor = "rgb(255, 255, 255)",
            countrycolor = "rgb(255, 255, 255)",
            countrywidth = 1,
            subunitwidth = 1
        ),
    )

fig = dict( data=cities, layout=layout )
iplot( fig, validate=False, filename='d3-bubble-map-populations' )
```

## Generate New Datasets off our previous data set
#### State_Awards_Fund will help us find our State-Model information
#### State_Awards will be used to find Average State Funding and Total State fund Funding
#### Fund_Awards will be used to report on what model types are funded and the average funding per model

```python
State_Awards_Fund=file.groupby(["State","Model Selected"])["Amount"].agg(['sum','count'])
State_Awards=file.groupby('State')['Amount'].agg(['sum','count'])
Fund_Awards=file.groupby('Model Selected')['Amount'].agg(['sum','count'])
```

```python
State_Awards_Fund.head()
```

```python
State_Awards.rename(columns ={'sum' : 'Award Total'}, inplace=True)
State_Awards.rename(columns ={'count' : 'Award Count'}, inplace=True)
State_Awards=State_Awards.reset_index()
State_Awards_Fund.rename(columns ={'sum' : 'Award Total'}, inplace=True)
State_Awards_Fund.rename(columns ={'count' : 'Award Count'}, inplace=True)
State_Awards_Fund=State_Awards_Fund.reset_index()
Fund_Awards.rename(columns ={'sum' : 'Award Total'}, inplace=True)
Fund_Awards.rename(columns ={'count' : 'Award Count'}, inplace=True)
Fund_Awards=Fund_Awards.reset_index()
```

```python
Fund_Awards.head()
```

```python
State_Awards['Award Count'] = State_Awards['Award Count'].astype('float64')
```

```python
# Create a trace
trace = go.Bar(
    x = State_Awards['State'],
    y = State_Awards['Award Total'],
    
)

data = [trace]

layout = go.Layout(
    title='Total Awarded by State',
    xaxis=dict(
        title='State',
        tickmode='linear',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
        )
    ),
    yaxis=dict(
        title='Award Total',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
        )
    )
)

fig = go.Figure(data=data, layout=layout)
# Plot and embed in ipython notebook!
iplot({'data': data, 'layout': layout}, filename='pandas-bar-chart-layout')

# or plot with: plot_url = py.plot(data, filename='basic-line')
```

```python
# Create a trace

trace = go.Bar(
    x = State_Awards['State'],
    y = State_Awards['Award Total']/State_Awards['Award Count'],
    
)

data = [trace]

layout = go.Layout(
    title='Average Awarded by State',
    xaxis=dict(
        title='State',
        tickmode='linear',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
        )
    ),
    yaxis=dict(
        title='Award Total',
        titlefont=dict(
            family='Courier New, monospace',
            size=18,
            color='#7f7f7f'
        )
    )
)

fig = go.Figure(data=data, layout=layout)
# Plot and embed in ipython notebook!
iplot({'data': data, 'layout': layout}, filename='pandas-bar-chart-layout')

# or plot with: plot_url = py.plot(data, filename='basic-line')
```
